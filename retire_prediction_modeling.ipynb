```python
data=pd.read_csv('e_employee_0810.csv')
```

```python
x_columns = [x for x in data.columns if x !='retire' and x !='EMPLOYEE_ID']
X = data[x_columns]
y = data['retire']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=82)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
from sklearn import cross_validation, metrics
```

```python
from sklearn.metrics import fbeta_score, make_scorer
ftwo_scorer = make_scorer(fbeta_score, beta=0.5)
param_test1 = {'n_estimators' : np.arange(100,1000,100)}
gsearch1 = GridSearchCV(estimator = RandomForestClassifier(min_samples_split=100,
                                  min_samples_leaf=20,max_depth=10,max_features='sqrt' ,random_state=10,class_weight={0:1,1:5}), 
                       param_grid = param_test1, scoring=ftwo_scorer,cv=5)
gsearch1.fit(X_train,y_train)
gsearch1.grid_scores_, gsearch1.best_params_, gsearch1.best_score_
```

```python
param_test2 = {'max_depth':np.arange(3,25,2), 'min_samples_split':np.arange(50,201,20)}
gsearch2 = GridSearchCV(estimator = RandomForestClassifier(n_estimators= 100, 
                                  min_samples_leaf=20,max_features='sqrt' ,oob_score=True, random_state=10,class_weight={0:1,1:5}),
   param_grid = param_test2, scoring=ftwo_scorer,iid=False, cv=5)
gsearch2.fit(X_train,y_train)
gsearch2.grid_scores_, gsearch2.best_params_, gsearch2.best_score_
```

```python
param_test3 = {'min_samples_split':np.arange(50,150,10), 'min_samples_leaf':np.arange(10,150,10)}
gsearch3 = GridSearchCV(estimator = RandomForestClassifier(n_estimators= 100, max_depth=15,
                                  max_features='sqrt' ,oob_score=True, random_state=10,class_weight={0:1,1:5}),
   param_grid = param_test3, scoring=ftwo_scorer,iid=False, cv=5)
gsearch3.fit(X_train,y_train)
gsearch3.grid_scores_, gsearch3.best_params_, gsearch3.best_score_
```

```python
param_test4 = {'max_features':np.arange(2,35,3)}
gsearch4 = GridSearchCV(estimator = RandomForestClassifier(n_estimators= 100, max_depth=15, min_samples_split=60,
                                  min_samples_leaf=10 ,oob_score=True, random_state=10,class_weight={0:1,1:5}),
   param_grid = param_test4, scoring=ftwo_scorer,iid=False, cv=5)
gsearch4.fit(X_train,y_train)
gsearch4.grid_scores_, gsearch4.best_params_, gsearch4.best_score_
```

```python
rf = RandomForestClassifier(n_estimators= 100,min_samples_split=60,
                                  min_samples_leaf=10,max_depth=15,max_features=8,random_state=10,oob_score=True,
                           class_weight={0:1,1:5})
rf.fit(X_train,y_train)
pred=rf.predict(X_train)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, pred)
```

```python
pred=rf.predict(X_test)
confusion_matrix(y_test, pred)
```
